```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
# Groups review rating by category and then fills them with the median of that category
```

```python
df.isnull().sum()
```

```python
#Convert all names to snake case for easy access in SQL and Python (Lowercase + underscores in-between)

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}) #rename the column to remove the () part

df.columns
```

```python
# Create a column 'age_group'
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels) #Sorts, Divides age by 4 quantiles and matches with labels
```

```python
df[['age','age_group']].head(10) #first 10 rows
```

```python
# Create column - purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Annually': 365,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all() # Checking if all values of these 2 columns are same or not
```

```python
#Since, both columns have the same values - promo_code_used column can be dropped

df = df.drop('promo_code_used', axis=1) # axis=1 means it is a column, axis=0 means row
df.columns # check columns again to see promo_code_used removed
```
